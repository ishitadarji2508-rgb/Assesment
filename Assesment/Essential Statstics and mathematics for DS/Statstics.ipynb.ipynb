# Theory Section

1. What is the difference between Descriptive and Inferential Statistics?

| Feature   | Descriptive Statistics                                          | Inferential Statistics                                                |
| --------- | --------------------------------------------------------------- | --------------------------------------------------------------------- |
| Purpose   | Describes and summarizes data                                   | Makes predictions or conclusions about a population based on a sample |
| Data Used | Entire dataset collected                                        | Sample data from a larger population                                  |
| Goal      | Understand what the data shows                                  | Generalize findings beyond the collected data                         |
| Methods   | Mean, Median, Mode, Range, Variance, Standard Deviation, Charts | Hypothesis Testing, Confidence Intervals, Regression, ANOVA           |
| Output    | Facts about the observed data                                   | Estimates, predictions, and decisions about a population              |


2. Define:
    Population
    Sample


1. Population

A population is the entire group of people, objects, or data that you want to study.

Example:
If you want to know the average marks of all students in a college, then all students in the college are the population.

2. Sample

A sample is a small part of the population selected for study.

Example:
If you select 100 students from the college and analyze their marks, those 100 students are the sample.

Easy Way to Remember
Population = Whole Group
Sample = Small Group Taken from the Whole Group

3. Define mean, median, and mode. How are they different from each other?

| Measure    | Definition                                      | Example                    |
| ---------- | ----------------------------------------------- | -------------------------- |
| **Mean**   | The average of all values                       | (10 + 20 + 30) ÷ 3 = 20    |
| **Median** | The middle value when data is arranged in order | 10, 20, 30 → Median = 20   |
| **Mode**   | The value that occurs most often                | 10, 20, 20, 30 → Mode = 20 |


Example

Data: 5, 8, 8, 10, 15

Mean = (5 + 8 + 8 + 10 + 15) ÷ 5 = 9.2
Median = 8 (middle value)
Mode = 8 (appears most often)

| Mean                                  | Median                            | Mode                                   |
| ------------------------------------- | --------------------------------- | -------------------------------------- |
| Uses all values                       | Uses only the middle value        | Uses the most frequent value           |
| Affected by extreme values (outliers) | Not much affected by outliers     | Not affected by outliers               |
| Best for numerical averages           | Best when data has extreme values | Best for finding the most common value |


4. What is a Random Variable? What are its types?


A Random Variable is a variable that takes numerical values based on the outcome of a random experiment.

Example:

Toss a coin:
Head = 1
Tail = 0

The variable X that represents the result is a random variable because its value depends on chance.

Types of Random Variables
1. Discrete Random Variable

A random variable that can take countable values.

Examples:

Number of heads in 3 coin tosses: 0, 1, 2, 3
Number of students in a classroom

Key Point: Values can be counted.

2. Continuous Random Variable

A random variable that can take any value within a range.

Examples:

Height of a person (170.2 cm, 170.25 cm, etc.)
Weight of a person
Temperature

Key Point: Values are measured and can have decimals.

Difference Between Discrete and Continuous
Discrete Random Variable	Continuous Random Variable
Countable values	Infinite values in a range
Obtained by counting	Obtained by measuring
Example: Number of cars	Example: Speed of a car
Easy Way to Remember
Discrete = Count (0, 1, 2, 3, ...)
Continuous = Measure (height, weight, time, temperature)

Example:

Number of books on a table → Discrete
Weight of those books → Continuous

5. Calculate:
o Mean, Median, Mode, Q1, Q3 (Quartiles)

```python
import pandas as pd

df = pd.read_csv("Student Attitude and Behavior.csv")

columns = ["Height(CM)", "Weight(KG)", "10th Mark", "12th Mark"]

for col in columns:
    print(f"\n{col}")
    print("Mean   :", df[col].mean())
    print("Median :", df[col].median())
    print("Mode   :", df[col].mode().tolist())
    print("Q1     :", df[col].quantile(0.25))
    print("Q3     :", df[col].quantile(0.75))
```

6. Find Skewness and Kurtosis. Interpret the results.


```python
cols = ["Height(CM)", "Weight(KG)", "10th Mark", "12th Mark"]

for col in cols:
    print("\n", col)

    skewness = df[col].skew()
    kurtosis = df[col].kurt()

    print("Skewness =", skewness)
    print("Kurtosis =", kurtosis)
```

7. Create a histogram and box plot for “10th Marks” and explain the shape of the
distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

print(df.columns)

marks = df["10th Mark"]

plt.figure(figsize=(6,4))
plt.hist(marks, bins=10)
plt.title("Histogram of 10th Marks")
plt.xlabel("10th Marks")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(4,4))
plt.boxplot(marks)
plt.title("Box Plot of 10th Marks")
plt.ylabel("10th Marks")
plt.show()
```
