Initialize file

```python
# File name
DATA_FILE = "data.txt"

# Create file if not exists
try:
    f = open(DATA_FILE, "x")
    f.close()
except:
    pass
```

Load data from file

```python
def load_data():

    stock = {}

    try:
        file = open(DATA_FILE, "r")

        for line in file:
            line = line.strip()

            if line != "":
                parts = line.split(",")

                name = parts[0]
                qty = int(parts[1])
                price = float(parts[2])

                stock[name] = {
                    "qty": qty,
                    "price": price
                }

        file.close()

    except:
        pass

    return stock
```

Save data to file

```python
def save_data(stock):

    file = open(DATA_FILE, "w")

    for fruit in stock:

        qty = stock[fruit]["qty"]
        price = stock[fruit]["price"]

        file.write(fruit + "," + str(qty) + "," + str(price) + "\n")

    file.close()
```

Manager functions

```python
def add_fruit(stock):

    print("\nADD FRUIT STOCK")

    name = input("Enter fruit name: ")

    qty = int(input("Enter quantity (kg): "))

    price = float(input("Enter price: "))

    stock[name] = {
        "qty": qty,
        "price": price
    }

    save_data(stock)

    print("Fruit added successfully")
```

```python
def view_fruit(stock):

    print("\nVIEW FRUIT STOCK")

    if len(stock) == 0:
        print("No stock available")

    else:
        for fruit in stock:

            print(fruit,
                  "Qty:", stock[fruit]["qty"],
                  "Price:", stock[fruit]["price"])
```

```python
def update_fruit(stock):

    print("\nUPDATE FRUIT")

    name = input("Enter fruit name: ")

    if name in stock:

        qty = int(input("Enter new quantity: "))
        price = float(input("Enter new price: "))

        stock[name]["qty"] = qty
        stock[name]["price"] = price

        save_data(stock)

        print("Fruit updated successfully")

    else:
        print("Fruit not found")
```

Customer function

```python
def customer_view(stock):

    print("\nAvailable Fruits")

    for fruit in stock:

        print(fruit,
              "Price:", stock[fruit]["price"])
```

```python
def buy_fruit(stock):

    name = input("Enter fruit name: ")

    if name not in stock:
        print("Fruit not available")
        return

    qty = int(input("Enter quantity: "))

    if qty > stock[name]["qty"]:
        print("Not enough stock")
        return

    total = qty * stock[name]["price"]

    stock[name]["qty"] -= qty

    save_data(stock)

    print("Purchase successful")
    print("Total price:", total)
```

Manager Menu

```python
def manager_menu():

    while True:

        stock = load_data()

        print("\nFRUIT MARKET MANAGER")
        print("1 Add Fruit")
        print("2 View Fruit")
        print("3 Update Fruit")
        print("4 Back")

        choice = input("Enter choice: ")

        if choice == "1":
            add_fruit(stock)

        elif choice == "2":
            view_fruit(stock)

        elif choice == "3":
            update_fruit(stock)

        elif choice == "4":
            break

        else:
            print("Invalid choice")
```

Customer Menu

```python
def customer_menu():

    while True:

        stock = load_data()

        print("\nCUSTOMER MENU")
        print("1 View Fruits")
        print("2 Buy Fruit")
        print("3 Back")

        choice = input("Enter choice: ")

        if choice == "1":
            customer_view(stock)

        elif choice == "2":
            buy_fruit(stock)

        elif choice == "3":
            break

        else:
            print("Invalid choice")
```

```python
while True:

    print("\nWELCOME TO FRUIT MARKET")
    print("1 Manager")
    print("2 Customer")
    print("3 Exit")

    role = input("Enter role: ")

    if role == "1":
        manager_menu()

    elif role == "2":
        customer_menu()

    elif role == "3":
        print("Program ended")
        break

    else:
        print("Invalid choice")
```
